*** Factorial Designs~Factorial Logistic & Factorial ANOVA with Python ***

** Factorial logistic regression **

```python
### Import Necessary Libraries

import numpy as np
import pandas as pd
```

```python
### Load Dataset and examine first 5 rows

url= "https://raw.githubusercontent.com/KenDaupsey/Factorial-Designs-Factorial-Logistic-Factorial-ANOVA-with-Python/main/hsb2%7Edata.csv"
df = pd.read_csv(url)

df.head()
```

```python
### Checking Categories of "female" variable

df['female'].value_counts()
```

```python
df.dtypes
```

```python
### Part 1: Encode “female” into numerical

from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode 'female' variable into numerical format
df['female_encoded'] = label_encoder.fit_transform(df['female'])

# Check the encoded values
print(df['female_encoded'].unique())

# Now 'female_encoded' can be used as the dependent variable in the logistic regression model
```

```python
df.dtypes
```

```python
df.head()
```

```python
##Checking Categories of new "female_encoded" variable

df['female_encoded'].value_counts()
```

```python
### Part 2: Fit Factorial Logistic Regression Model

#Factorial logistic regression
import statsmodels.api as sm

# Assuming df is your DataFrame
# Fit logistic regression model
formula = 'female_encoded ~ C(prog) + C(schtyp) + C(prog):C(schtyp)'
model = sm.Logit.from_formula(formula, data=df)
result = model.fit()

# Print the summary
print(result.summary())
```

```python
### Part 3: Hypothesis Tests on Specific Contrast

from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode 'female' variable into numerical format
df['female_encoded'] = label_encoder.fit_transform(df['female'])

# Check the encoded values
print(df['female_encoded'].unique())

# Now 'female_encoded' can be used as the dependent variable in the logistic regression model

# Now perform the Wald test with the corrected dependent variable
formula = 'female_encoded ~ C(prog) + C(schtyp) + C(prog):C(schtyp)'
model = sm.Logit.from_formula(formula, data=df)
result = model.fit()

# Perform a hypothesis test for 2.prog vs. 1.prog
contrast_matrix = [[0, 1, 0, 0, 0, 0]]  # Contrast for 2.prog vs. 1.prog
hypothesis_test_2_vs_1 = result.wald_test(contrast_matrix, scalar=True)
# Perform a hypothesis test for 3.prog vs. 1.prog
contrast_matrix = [[0, 0, 1, 0, 0, 0]]  # Contrast for 3.prog vs. 1.prog
hypothesis_test_3_vs_1 = result.wald_test(contrast_matrix, scalar=True)

# Print the results
print("Hypothesis test for 2.prog vs. 1.prog:")
print(hypothesis_test_2_vs_1)

print("\nHypothesis test for 3.prog vs. 1.prog:")
print(hypothesis_test_3_vs_1)
```

```python
# Part 4: Hypothesis Test for Interaction Effect

from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode 'female' variable into numerical format
df['female_encoded'] = label_encoder.fit_transform(df['female'])

# Check the encoded values
print(df['female_encoded'].unique())

# Now 'female_encoded' can be used as the dependent variable in the logistic regression model

# Now perform the Wald test with the corrected dependent variable
formula = 'female_encoded ~ C(prog) * C(schtyp)'
model = sm.Logit.from_formula(formula, data=df)
result = model.fit()

# Define the hypothesis matrix for prog#schtyp
hypothesis_matrix = [[0, 0, 0, 0, 0, 1]]

# Perform the Wald test with the corrected dependent variable
wald_test_result = result.wald_test(hypothesis_matrix, scalar=False)

# Print the result
print("Wald Test for prog#schtyp:")
print(wald_test_result)
```

*** Factorial ANOVA ***

```python
### Factorial ANOVA: Math Scores by SES and Program

import statsmodels.api as sm
from statsmodels.formula.api import ols

# Assuming the dataset is already loaded into a pandas DataFrame called 'df'
# And the columns are named 'ses', 'prog', and 'math'

# Fit the factorial ANOVA model
formula = 'math ~ C(ses) * C(prog)'
model = ols(formula, data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display the ANOVA table
print(anova_table)
```

```python
### Factorial ANOVA + post-calculation command
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Assuming 'model' is the fitted factorial ANOVA model from the previous code

# Perform Tukey-Kramer post hoc test
tukey_results = pairwise_tukeyhsd(df['math'], df['ses'].astype(str) + '_' + df['prog'].astype(str), alpha=0.05)

# Display the post hoc test results
print(tukey_results.summary())
```
